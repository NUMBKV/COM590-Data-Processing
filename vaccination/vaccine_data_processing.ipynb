```python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy
import requests
import pandas as pd
from sodapy import Socrata
import os



# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cdc.gov",None)

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get_all("8xkx-amqh", date='2022-03-01')

# Convert to pandas DataFrame
df = pd.DataFrame.from_records(results)
display(df)
```

```python
df_group = df.sort_values(by='recip_state')
display(df_group)
df_group.to_csv("vaccine_county_state.csv")
```

```python
df_dose1 = df.copy()
df_dose1 = df_dose1.drop(df.columns[1:3], axis=1) 
df_dose1 = df_dose1.drop(df.columns[16:], axis=1) 
display(df_dose1)
df_group.to_csv("first_dose_county_state.csv")
```
